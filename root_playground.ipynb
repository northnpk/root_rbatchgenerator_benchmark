```python
def join_time(row):
    row['Init time'] = row[['Init time1', 'Init time2', 'Init time3', 'Init time4']].to_list()
    row['Iteration time'] = row[['Iteration time1', 'Iteration time2', 'Iteration time3', 'Iteration time4']].to_list()
    row['Total time'] = row[['Total time1', 'Total time2', 'Total time3', 'Total time4']].to_list()
    return row
```

```python
df = df.apply(join_time, axis=1)
```

```python
df.columns
```

```python
df
```

```python
df = df[['Log type', 'Tools', 'Init time', 'Iteration time', 'Total time']]
```

```python
def avg(lst):
    return sum(lst) / len(lst)

def calc_error(row):
    i_t = row['Init time']
    iter_t = row['Iteration time']
    total_t = row['Total time']

    row['Init avg'] = avg(i_t)
    row['Init min'] = avg(i_t)-min(i_t)
    row['Init max'] = max(i_t)-avg(i_t)

    row['Iter avg'] = avg(iter_t)
    row['Iter min'] = avg(iter_t)-min(iter_t)
    row['Iter max'] = max(iter_t)-avg(iter_t)

    row['Total avg'] = avg(total_t)
    row['Total min'] = avg(total_t)-min(total_t)
    row['Total max'] = max(total_t)-avg(total_t)

    return row
```

```python
df = df.apply(calc_error, axis=1)
```

```python
df.columns
```

```python
def naming(row):
    name = row['Log type'] + row['Tools']
    return name
```

```python
df['x_label'] = df.apply(naming, axis=1)
```

```python
import awkward as ak
def _pad(a, maxlen, value=0, dtype='float32'):
    if isinstance(a, np.ndarray) and a.ndim >= 2 and a.shape[1] == maxlen:
        return a
    elif isinstance(a, ak.Array):
        if a.ndim == 1:
            a = ak.unflatten(a, 1)
        a = ak.fill_none(ak.pad_none(a, maxlen, clip=True), value)
        return ak.values_astype(a, dtype)
    else:
        x = (np.ones((len(a), maxlen)) * value).astype(dtype)
        for idx, s in enumerate(a):
            if not len(s):
                continue
            trunc = s[:maxlen].astype(dtype)
            x[idx, :len(trunc)] = trunc
        return x
```
